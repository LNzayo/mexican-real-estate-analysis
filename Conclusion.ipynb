# Hands-on Data Science in the Mexican Real Estate Market

## 📋 Executive Summary

This project analyzes the Mexican real estate market using a dataset of **21,000+ properties** from Properati.com to determine whether property prices are driven more by **location** or **property size**. Through comprehensive data cleaning, exploratory data analysis, and correlation analysis, the project reveals distinct regional dynamics across Mexican states, providing actionable insights for buyers, sellers, and real estate professionals.

## 🎯 Project Objectives

| Objective | Description |
|-----------|-------------|

## 🏆 Conclusion

### Summary of Findings

| Finding | Implication |
|---------|-------------|
| **Location is dominant in CDMX** | Size alone does not explain prices |
| **Size drives prices in Morelos** | Strong positive correlation |
| **Regional variation exists** | No one-size-fits-all answer |
| **Price per m² is key metric** | Better than average price |

### Final Answer

**Does location or size drive prices more?**

> **It depends on the region**

**Key Takeaway:** While both factors matter, their relative importance varies significantly across Mexico's real estate market:

- **Mexico City:** Location is the dominant factor
- **Morelos:** Property size is the primary driver
- **Other states:** Mixed dynamics with varying influence
